```python
#extract to csv
from bs4 import BeautifulSoup  
import requests
import csv

page = requests.get('https://aeonmall-bsdcity.com/shopping.php')
raw_html = page.text    # declare the raw_html var

soup = BeautifulSoup(raw_html, 'html.parser')  # parse the html

T = [["AEON MALL BSD CITY"]] #Title

links = map(lambda link: link['href'], soup.find_all('a', href=True)) # links

with open("aeon_mall_bsd_city.csv","w",newline="") as f:    
    cw=csv.writer(f, quoting=csv.QUOTE_ALL)                          
    cw.writerows(T)                             
    cw.writerow(links)                         

f.close()
```
